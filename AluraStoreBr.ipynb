### Importação dos dados



````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
````

#1. Análise do faturamento


````python
Mf = 0
for i in range (4):
  if i == 0:
    ft_l1 = loja['Preço'].sum()
    print(f"O faturamento total da Loja 1 foi de {ft_l1:,.2f}")
  elif i == 1:
    ft_l2 =  loja2['Preço'].sum()
    print(f"O faturamento total da Loja 2 foi de {ft_l2:,.2f}")
  elif i == 2:
    ft_l3 =  loja3['Preço'].sum()
    print(f"O faturamento total da Loja 3 foi de {ft_l3:,.2f}")
  elif i == 3:
    ft_l4 =  loja4['Preço'].sum()
    print(f"O faturamento total da Loja 4 foi de {ft_l4:,.2f}")

def menor_faturamento(Faturamento_total_l1,Faturamento_total_l2,Faturamento_total_l3, Faturamento_total_l4):
  if Faturamento_total_l1 < Faturamento_total_l2 and Faturamento_total_l1 < Faturamento_total_l3 and Faturamento_total_l1 < Faturamento_total_l4:
    mf = Faturamento_total_l1
    print(f"O menor faturamento foi da loja 1 e foi de R$ {mf:,.2f}")
  elif Faturamento_total_l2 < Faturamento_total_l1 and Faturamento_total_l2 < Faturamento_total_l3 and Faturamento_total_l2 < Faturamento_total_l4:
    mf = Faturamento_total_l2
    print(f"O menor faturamento foi da loja 2 e foi de R$ {mf:,.2f}")
  elif Faturamento_total_l3 < Faturamento_total_l1 and Faturamento_total_l3 < Faturamento_total_l2 and Faturamento_total_l3 < Faturamento_total_l4:
    mf = Faturamento_total_l3
    print(f"O menor faturamento foi da loja 3 e foi de R$ {mf:,.2f}")
  elif Faturamento_total_l4 < Faturamento_total_l1 and Faturamento_total_l4 < Faturamento_total_l2 and Faturamento_total_l4 < Faturamento_total_l3:
    mf = Faturamento_total_l4
    print(f"O menor faturamento foi da loja 4 e foi de R$ {mf:,.2f}")

fat_total = (ft_l1 + ft_l2 + ft_l3 + ft_l4)

menor_faturamento(ft_l1,ft_l2,ft_l3,ft_l4)
print(f"O faturamento total foi de R$ {fat_total:,.2f}")
````

# 3. Média de Avaliação das Lojas

# 4. Produtos Mais e Menos Vendidos

````python
from collections import Counter

def maisVendido_MenosVendido(nome_loja, l):
  qt_produtos = Counter(l['Produto'])
  tm = len(qt_produtos)
  mais_v = qt_produtos.most_common(1)[0]
  menos_v = qt_produtos.most_common(tm)[tm - 1]
  print(f'o produto mais vendido da {nome_loja} é o {mais_v}')
  print(f'o produto menos vendido da {nome_loja} é o {menos_v}')


maisVendido_MenosVendido('Loja 1', loja)
maisVendido_MenosVendido('Loja 2', loja2)
maisVendido_MenosVendido('Loja 3', loja3)
maisVendido_MenosVendido('Loja 4', loja4)
````
